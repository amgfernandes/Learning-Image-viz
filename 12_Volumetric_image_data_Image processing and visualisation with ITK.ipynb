# Image processing and visualisation with ITK
See also: https://itkpythonpackage.readthedocs.io/en/master/Quick_start_guide.html

```python
import itk

input_filename = "/Users/fernandes/Documents/Data/MR.dcm"

image = itk.imread(input_filename)

print(image)
```

```python
print(image.GetSpacing())
```

```python
voxel_size_z = image.GetSpacing()[0]

print(voxel_size_z)
```

```python
# applying a filter to an image
median_filtered = itk.median_image_filter(image, radius=2)
```

```python
# Mixing ITK with numpy and matplotlib: e.g. for visualisation
import numpy as np
import matplotlib.pyplot as plt

# convert itk image to a numpy array
np_view = itk.array_view_from_image(median_filtered)

# show the image using matplotlib
plt.imshow(np_view[0])
plt.show()
```
