```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
from skimage import data, io
from skimage import color, morphology


'''venv: bioimage'''
```

```python
image = color.rgb2gray(data.hubble_deep_field())[:600, :600]


# Gamma adjust to see better
image = exposure.adjust_gamma(image, 0.8)

footprint = morphology.disk(1)
res = morphology.white_tophat(image, footprint)

fig, ax = plt.subplots(ncols=3, figsize=(20, 8))
ax[0].set_title('Original')
ax[0].imshow(image, cmap='gray')
ax[1].set_title('White tophat')
ax[1].imshow(res, cmap='gray')
ax[2].set_title('Complementary')
ax[2].imshow(image - res, cmap='gray')
```

```python

fig, ax = plt.subplots(ncols=2, figsize=(20, 8))
ax[0].set_title('Original')
ax[0].imshow(image, cmap='gray')
ax[1].set_title('Complementary white hat')
ax[1].imshow(image - res, cmap='gray')
```

```python
footprint = morphology.disk(1)
res = morphology.opening(image, footprint)

fig, ax = plt.subplots(ncols=2, figsize=(20, 8))
ax[0].set_title('Original')
ax[0].imshow(image, cmap='gray')
ax[1].set_title('opening only')
ax[1].imshow(res, cmap='gray')
```
